```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/List_of_colloquial_names_for_universities_and_colleges_in_the_United_States')
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
AZ = soup.find(id ='content')
```

```python
items =AZ.find_all('ul')
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
A = items[1].get_text()
A = A.split('\n')
A = pd.DataFrame({'Names':A})
A1 = A.Names.apply(lambda x: pd.Series(str(x).split("–")))
A1.columns=['Acronym','College','College2']
A1.head()
```
